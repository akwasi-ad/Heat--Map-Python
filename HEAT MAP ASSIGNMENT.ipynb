```python
import pandas as pd 
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("gapminder-FiveYearData.csv")
df
```

```python
df.head()
```

```python
gap_minder = df.pivot_table(index='continent',columns='year',values='lifeExp',aggfunc='sum')  
gap_minder
```

```python
plt.figure(0, figsize=(13, 13))
display_gap_minder=sns.heatmap(gap_minder,annot=True,fmt='.2f')
display_gap_minder
plt.savefig('display_gap_minder.png')
```
